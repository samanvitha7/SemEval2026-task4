```python
import json
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
with open("track_b_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

story_ids = []
stories = []

for item in data:
    story_ids.append(item["story_id"])
    stories.append(item["story_text"])
```

```python
#text cleaning
def clean_text(text):
    text = text.lower()
    text = text.replace("\n", " ")
    text = " ".join(text.split())
    return text

stories = [clean_text(s) for s in stories]
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(
    stories,
    batch_size=32,
    convert_to_numpy=True,
    normalize_embeddings=True
)
```

```python
from sentence_transformers.util import cos_sim

print(cos_sim(embeddings[0], embeddings[0]))   # ~1.0
print(cos_sim(embeddings[0], embeddings[1]))   # lower value
```

```python
output = []

for sid, emb in zip(story_ids, embeddings):
    output.append({
        "story_id": sid,
        "embedding": emb.tolist()
    })

with open("track_b_baseline_embeddings.json", "w") as f:
    json.dump(output, f)
```
